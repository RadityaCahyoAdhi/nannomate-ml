# Import library

```python
import pandas as pd
import numpy as np
from sklearn import preprocessing
from collections import Counter
from imblearn.over_sampling import RandomOverSampler
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import classification_report
from jcopml.utils import save_model
from jcopml.utils import load_model
```

# Data preprocessing

```python
# read data
df = pd.read_csv('Dataset_Discoaster(Self-Rev).csv')
df.head()
```

```python
df.shape
```

```python
# feautures encoding
ordl_encoder = preprocessing.OrdinalEncoder(dtype=int)
ordl_encoder.fit(df[['Bercabang', 'Knob', 'Bentuk_Lengan', 'Ujung_Lengan']])
df[['Bercabang', 'Knob', 'Bentuk_Lengan', 'Ujung_Lengan']] = ordl_encoder.transform(df[['Bercabang', 'Knob', 'Bentuk_Lengan', 'Ujung_Lengan']])
```

```python
df.head()
```

```python
# target label encoding
lbl_encoder = preprocessing.LabelEncoder()
lbl_encoder.fit(df['Spesies'])
df['Spesies'] = lbl_encoder.transform(df['Spesies'])
```

```python
df.head()
```

```python
df.shape
```

```python
# check dataset balance
sorted(df['Spesies'].unique())
```

```python
print(sorted(Counter(df['Spesies']).items()))
```

```python
df['Spesies'].value_counts().describe()
```

```python
# features and targets split
X = df[['Jumlah_Lengan', 'Bercabang', 'Knob', 'Bentuk_Lengan', 'Ujung_Lengan']]
X
```

```python
y = df['Spesies']
y
```

```python
# random oversampling
ros = RandomOverSampler(random_state=0)
X_resampled, y_resampled = ros.fit_resample(X, y)
```

```python
# repeat check dataset balance
X_resampled
```

```python
y_resampled
```

```python
print(sorted(Counter(y_resampled).items()))
```

```python
y_resampled.value_counts().describe()
```

Train Test Split

```python
X_train, X_test, y_train, y_test = train_test_split(X_resampled, y_resampled, test_size=0.2, random_state=1945, stratify=y_resampled)
```

```python
print(sorted(Counter(y_train).items()))
```

```python
y_train.value_counts().describe()
```

```python
print(sorted(Counter(y_test).items()))
```

```python
y_test.value_counts().describe()
```

# Algorithm Selection Test

```python
# Support Vector Machine
svm = SVC()
svm_cv_results = cross_val_score(svm, X_train, y_train, cv=5, verbose=3)
```

```python
print("accuracy: {}, std: {}".format(svm_cv_results.mean(), svm_cv_results.std()))
```

```python
# Random Forest
random_forest = RandomForestClassifier()
random_forest_cv_results = cross_val_score(random_forest, X_train, y_train, cv=5, verbose=3)
```

```python
print("accuracy: {}, std: {}".format(random_forest_cv_results.mean(), random_forest_cv_results.std()))
```

```python
# neural network
neural_network = MLPClassifier()
neural_network_cv_results = cross_val_score(neural_network, X_train, y_train, cv=5, verbose=3)
```

```python
print("accuracy: {}, std: {}".format(neural_network_cv_results.mean(), neural_network_cv_results.std()))
```

```python
# print each model accuracy (std)
print("svm(svc): {} ({})".format(svm_cv_results.mean(), svm_cv_results.std()))
print("random forest: {} ({})".format(random_forest_cv_results.mean(), random_forest_cv_results.std()))
print("neural network: {} ({})".format(neural_network_cv_results.mean(), neural_network_cv_results.std()))
```

# Create Model

```python
# number of trees in random forest
n_estimators = [50, 100, 150, 200, 250, 300, 350, 400, 450, 500]
# split criterion
criterion = ["gini", "entropy"]
# maximum number of levels in tree
max_depth = [5, 10, 15, 20, 25, 30, None]
# max considered features per split
max_features = ["auto", "log2", None]
# min samples per leaf node
min_samples_leaf = [1, 2, 3, 4, 5]
# create the params grid
param_grid = {'n_estimators': n_estimators,
              'criterion': criterion,
               'max_depth': max_depth,
             'max_features': max_features,
             'min_samples_leaf': min_samples_leaf}
```

```python
# GridSearchCV fit
model = RandomForestClassifier(n_jobs=2, random_state=1)

grid = GridSearchCV(estimator=model, param_grid=param_grid, scoring="accuracy", cv=5, verbose=3)
grid_result = grid.fit(X_train, y_train)
```

```python
# GridSearchCV result
print("Best: {} using {}".format(round(grid_result.best_score_, 6), grid_result.best_params_)) 

means = grid_result.cv_results_['mean_test_score']
stds = grid_result.cv_results_['std_test_score']
params = grid_result.cv_results_['params']

for mean, stdev, param in zip(means, stds, params):
    print("%f (%f) with: %r" % (mean, stdev, param))
```

```python
# fit final model
# final_model = RandomForestClassifier(n_estimators=450, max_depth=None)
final_model = RandomForestClassifier(criterion='entropy', max_depth=10, max_features='auto', min_samples_leaf= 2, n_estimators=300, n_jobs=2, random_state=1)
final_model.fit(X_train, y_train)

y_pred = final_model.predict(X_test)
```

```python
# train test accuracy score
print("train accuracy: {}, test accuracy: {}".format(round(final_model.score(X_train, y_train), 3), round(final_model.score(X_test, y_test), 3)))
```

```python
# classification report
print(classification_report(y_test, y_pred, digits=3))
```
